RESPOSTA DA QUESTÃO 2


```python
#primeiro tem que importar o classificador de face
detector_facial = cv2.CascadeClassifier('/content/haarcascade_frontalface_default.xml')
```

```python
#Importando a imagem utilizada para detecção de face
image_face = cv2.imread('/content/caras.jpg')

#convertendo a imagem para escala de cinza e 
image_gray = cv2.cvtColor(image_face, cv2.COLOR_BGR2GRAY)
detector = detector_facial.detectMultiScale(image_gray, scaleFactor=1.1, minNeighbors=3, 
                                             minSize=(32,32), maxSize=(400,400))

#Criando um for através do detector para desenhar as bbox e printar seu tamanho
for (x, y, w, h) in detector:
  print(w, h)
  cv2.rectangle(image_face, (x, y), (x + w, y + h), (400,255,0), 2)


#Mostrar a imagem final com as bbox
cv2_imshow(image_face)
```
